```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties

# Li 实验的正态分布参数
mu = 0.75  # 均值
sigma = 0.13  # 标准差
min_val, max_val = 0.52, 1.19  # 限定范围

# 生成符合 Li 实验正态分布的数据
np.random.seed(42)  # 设定随机种子，保证结果可复现
samples = np.random.normal(mu, sigma, 10000)  # 生成 10000 个样本
samples = samples[(samples >= min_val) & (samples <= max_val)]  # 仅保留范围内的数据

# 计算经验累积分布（ECDF）
plt.figure(figsize=(6, 5))
sns.ecdfplot(samples, color="blue", label="Li.et N(0.75,0.13^2)")

# 标注本模型计算值
plt.axvline(x=0.79, color="red", linestyle="--", linewidth=2, label="this model")
plt.text(0.79+ 0.02, 0.55, "0.79 m/s", color="red", fontsize=12,ha='left', va='center')


# 设置标题和坐标轴
plt.xlabel("velocity (m/s)")
plt.ylabel("Cumulative probability")
plt.title("Cumulative distribution of pedestrian underwater free speed (ECDF)")

plt.legend()
plt.show()
```
